```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import KFold, train_test_split
```

```python
data = pd.read_parquet('d:/BankCustomer/data/bank_v3.parquet')
df = data.copy()
```

```python
df.head().T
```

<center><h1>Split Data</h1></center>

```python
splitter = KFold(n_splits=5, shuffle=True, random_state=33)
folds = list(splitter.split(df))
```

```python
folds[0]
```

```python
folds[2]
```

```python
df.loc[folds[0][0]]
```

```python
d_folds = {}
fold = 0
for train_index, test_index in folds:
    d_folds[f'fold_{fold}_train'] = train_index
    d_folds[f'fold_{fold}_test'] = test_index
    fold += 1
```

```python
d_folds
```

```python
np.savez('d:/BankCustomer/data/folds.npz', **d_folds)
```
